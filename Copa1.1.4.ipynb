```python
import pandas as pd
import numpy as np

g = [['RUS', 'KSA', 'EGY', 'URU'], ['MAR', 'IRN', 'POR', 'ESP'], ['FRA', 'AUS', 'PER', 'DEN'], ['ARG', 'ISL', 'CRO', 'NGA'], ['BRA', 'SUI', 'CRC', 'SRB'], ['SWE', 'MEX', 'GER', 'KOR'], ['ENG', 'PAN', 'BEL', 'TUN'], ['JPN', 'SEN', 'POL', 'COL']]
e = ['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H']
#cod = 1

jogos = pd.read_json('https://t.co/5kiIGpo1vg', orient='columns')


def transf_pd(table, data):
    a = table.loc[:,data]
    lista = []
    for i in a:
        lista.append(i)
    return pd.DataFrame(lista)


away_pd = transf_pd(jogos,'away_team')
home_pd = transf_pd(jogos,'home_team')


def grupo(grupo):
    global cod
    print('Grupo', e[grupo])
    print(f" {'ID'} {'Data'}        {'Hora'}   {'Score'}           {'Estádio'}")
    for cont in range(0,63):
        if home_pd.loc[cont,'code'] in g[grupo] and jogos.loc[cont,'stage_name'] == 'First stage':
            data = str(jogos.loc[cont, 'datetime'])
            print(f' {cont+1:>2} {data[:11]} {data[11:16]}', end="")
            print(' ', home_pd.loc[cont, 'code'], home_pd.loc[cont, 'goals'], 'x', away_pd.loc[cont, 'goals'], away_pd.loc[cont, 'code'], end="   ")
            print(jogos.loc[cont, 'location'])
        cont += 1
    print()
 

def mata(fase):
    global cod
    print(fase)
    print(f" {'ID'} {'Data'}        {'Hora'}   {'Score'}         {'Penaltis'}   {'Estádio'}")
    for cont in range(0,64):
        if jogos.loc[cont,'stage_name'] == fase:
            data = str(jogos.loc[cont, 'datetime'])
            print('', cont+1, data[:11], data[11:16], end="")
            if home_pd.loc[cont, 'goals'] == away_pd.loc[cont, 'goals']:
                print(' ', home_pd.loc[cont, 'code'], home_pd.loc[cont, 'goals'], 'x', away_pd.loc[cont, 'goals'], away_pd.loc[cont, 'code'], '(', home_pd.loc[cont, 'penalties'],'x',away_pd.loc[cont, 'penalties'],')', end="  ")
            else:
                print(' ', home_pd.loc[cont, 'code'], home_pd.loc[cont, 'goals'], 'x', away_pd.loc[cont, 'goals'], away_pd.loc[cont, 'code'], end="            ")
            print(jogos.loc[cont, 'location'])
        cont += 1
    print()
    
            
for i in range(0, 8):
    grupo(i)

mata('Round of 16')
mata('Quarter-finals')
mata('Semi-finals')
mata('Play-off for third place')
mata('Final')

```

```python
away_stats = transf_pd(jogos,'away_team_statistics')
home_stats = transf_pd(jogos,'home_team_statistics')

def stats(num):
    num = num - 1
    home = home_stats.loc[num,'country']
    away = away_stats.loc[num,'country']
    print(f"{home:^15} {' '*30} {away:^15}")
    print(f"{home_stats.loc[num,'attempts_on_goal']:^15} {'CHUTES A GOL':^30} {away_stats.loc[num,'attempts_on_goal']:^15}")
    print(f"{home_stats.loc[num,'ball_possession']:^15} {'POSSE DE BOLA':^30} {away_stats.loc[num,'ball_possession']:^15}")
    print(f"{home_stats.loc[num,'corners']:^15} {'ESCANTEIOS':^30} {away_stats.loc[num,'corners']:^15}")
    if np.isnan(home_stats.loc[num,'fouls_committed']) and np.isnan(away_stats.loc[num,'fouls_committed']):
        print(f"{'0':^15} {'FALAS COMETIDAS':^30} {'0':^15}")
    elif np.isnan(home_stats.loc[num,'fouls_committed']):   
        print(f"{'0':^15} {'FALAS COMETIDAS':^30} {int(away_stats.loc[num,'fouls_committed']):^15}")
    elif np.isnan(home_stats.loc[num,'fouls_committed']):
        print(f"{int(home_stats.loc[num,'fouls_committed']):^15} {'FALAS COMETIDAS':^30} {'0':^15}")
    else:
        print(f"{int(home_stats.loc[num,'fouls_committed']):^15} {'FALAS COMETIDAS':^30} {int(away_stats.loc[num,'fouls_committed']):^15}")
    print(f"{home_stats.loc[num,'on_target']:^15} {'CHUTES A GOL':^30} {away_stats.loc[num,'on_target']:^15}")
    print(f"{home_stats.loc[num,'offsides']:^15} {'IMPEDIMENTOS':^30} {away_stats.loc[num,'offsides']:^15}")
    print(f"{home_stats.loc[num,'passes_completed']:^15} {'PASSES CORRETOS':^30} {away_stats.loc[num,'passes_completed']:^15}")
    print(f"{(home_stats.loc[num,'num_passes'] - home_stats.loc[num,'passes_completed']):^15} {'PASSES ERRADOS':^30} {(away_stats.loc[num,'num_passes'] - away_stats.loc[num,'passes_completed']):^15}")
    print(f"{home_stats.loc[num,'yellow_cards']:^15} {'CARTÕES AMARELOS':^30} {away_stats.loc[num,'yellow_cards']:^15}")
    print(f"{home_stats.loc[num,'red_cards']:^15} {'CARTÕES VERMELHOS':^30} {away_stats.loc[num,'red_cards']:^15}")
    
#stats(34)
print(home_stats.head())
```

```python
goal=[]
for c in range(0, 64):
    home_event = pd.DataFrame(jogos.loc[c,'home_team_events'])
    away_event = pd.DataFrame(jogos.loc[c,'away_team_events'])
    for i in range(0, len(home_event)):
        if home_event.loc[i,'type_of_event']=='goal' or home_event.loc[i,'type_of_event']=='goal-penalty':
            goal.append(home_event.loc[i,'player'])
    for i in range(0, len(away_event)):    
        if away_event.loc[i,'type_of_event']=='goal'or away_event.loc[i,'type_of_event']=='goal-penalty':
            goal.append(away_event.loc[i,'player'])   

goals = dict()

for l in goal:
        if l in goals.keys():
            goals[l] += 1
        else:
            goals[l] = 1
            
goals_counts_list = []
for a, b in goals.items():
    goals_counts_list.append([a, b])
    
goals_counts_list.sort(key=lambda x: x[1], reverse=True)       

print(f'{"Artilheiros":^20}')
for i in range(0, 15):
    print(f'{goals_counts_list[i][0]:>18} {goals_counts_list[i][1]:>2}')            
```
